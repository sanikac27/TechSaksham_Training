```python
x1_axis=[11,12,13]
y1_axis=[21,32,45]

x2_axis=[9,4,1]
y2_axis=[32,4,5]
```

```python
plt.subplot(2,1,1) #how_many_row how_many_col currently_plot_no  #(1,2,2)
plt.title('Exponential')
plt.plot(x1_axis,y1_axis,'r')

plt.subplot(2,1,2)     #(2,1,2)
plt.title('linear Values')
plt.plot(x2_axis,y2_axis,'g')

plt.tight_layout()
```

```python
import pandas as pd
import seaborn as sns
dataset= pd.read_csv('data.csv')
```

```python
sns.relplot(x='hours' ,y='marks',hue='age',data=dataset)  #hue gives color combination to graph
```

```python
sns.relplot(x='hours' ,y='marks',data=dataset)
```

```python
sns.relplot(x='hours' ,y='marks',hue='internet',data=dataset) # o , 1 value like color
```

# statical line plot

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset,
           kind='line',style='age')
```

```python
sns.relplot(x='hours',y='marks',hue='age',data=dataset,
           col='internet')
```

```python
sns.relplot(x='hours',y='marks',hue='internet',data=dataset,
           col='age')
```

# Data Distribution - Histogram

```python
sns.displot(dataset,x="age",binwidth=0.5,hue="internet")
```

# Data Distribution - Multiple

```python
sns.displot(dataset,x='marks',col='internet')
```

# Data Distribution - bivarient

```python
sns.displot(dataset,x='age',y='marks',hue='internet')
```

```python
sns.displot(dataset,x='age',y='marks',
            kind='kde',hue='internet')
```

# Data Distribution -Pair plot

```python
sns.pairplot(dataset)
```

# Categorical data - scatter plot

```python
sns.catplot(x='age',y='marks', data=dataset,hue='internet')
```

# categorical data- box plot

```python
sns.catplot(x='age',y='marks',kind='box',data=dataset,hue='internet')
```

# Regression based plot

```python
sns.regplot(x='marks',y='hours',data=dataset)
```

```python
sns.lmplot(x='marks',y='hours',data=dataset,
          hue='internet',markers=['o','x'],
          palette='Set1')
```
